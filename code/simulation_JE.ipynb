```python
from makeTempHistogramPlot import *
from generateInfo import *

# Global constants
G_GRAV = 4.54E-48 #units of Mpc^3 Msun^-1 s^-2
KM_TO_MPC = 3.2E-20
HUBBLE = 0.70
E_UNITS = 1.0E-36
L_UNITS = 1.0

# Global filepaths
#data_path="/home/jeforero/Dropbox/"
#data_path="/home/forero/Dropbox/"
data_path="/Users/forero/Dropbox/"
```

```python
# The reference is van der Marel et al. ApJ,753:8
V_M31 = array([66.1,-76.3,45.1]) #km/s
V_M31_sigma = array([26.7, 19.0, 26.5])

V_M31_sigma = V_M31_sigma/4

R_M31 = array([-378.9,612.7,-283.1]) # kpc
R_M31_sigma = abs(R_M31/20.0) #This is an approximated value, not the actual observational uncertainty

print R_M31_sigma, sqrt(sum(R_M31*R_M31))

R_M31 = R_M31 / 1000.0 # in Mpc
R_M31_sigma = R_M31_sigma /1000.0

Mass_Total = 3.14E12 # in units of Msun
Mass_Total_sigma = 0.58E12
```

```python
#running the montecarlo for the energy and the angular momentum
n_points = 1000000
V_M31_MonteCarlo = zeros((n_points,3))
R_M31_MonteCarlo = zeros((n_points,3))

Mass_M31_MonteCarlo = random.normal(Mass_Total,Mass_Total_sigma, n_points)
for i in range(3):
    V_M31_MonteCarlo[:,i] = random.normal(V_M31[i],V_M31_sigma[i],n_points)
    R_M31_MonteCarlo[:,i] = random.normal(R_M31[i],R_M31_sigma[i],n_points)

J = cross(R_M31_MonteCarlo, V_M31_MonteCarlo) # array of angular momentum (vector)
J_norm = sqrt(J[:,0]**2 + J[:,1]**2 + J[:,2]**2) # and the norm

R_M31_MonteCarlo_norm  = sqrt(R_M31_MonteCarlo[:,0]**2 + R_M31_MonteCarlo[:,1]**2 + R_M31_MonteCarlo[:,2]**2)
E_kin = 0.5*(V_M31_MonteCarlo[:,0]**2 + V_M31_MonteCarlo[:,1]**2 + V_M31_MonteCarlo[:,2]**2)
E_kin = E_kin * KM_TO_MPC * KM_TO_MPC
E_pot = -G_GRAV * Mass_M31_MonteCarlo / R_M31_MonteCarlo_norm

E_kin = E_kin/E_UNITS
E_pot = E_pot/E_UNITS

E = E_kin + E_pot # array of mechanical energy
print amin(J_norm), amax(J_norm)
print amin(E), amax(E)
print amin((2.0*E_UNITS*E_kin/KM_TO_MPC/KM_TO_MPC)), amax((2.0*E_UNITS*E_kin/KM_TO_MPC/KM_TO_MPC)),
print std((2.0*E_UNITS*E_kin/KM_TO_MPC/KM_TO_MPC))/mean((2.0*E_UNITS*E_kin/KM_TO_MPC/KM_TO_MPC)), 22.0/116.0, 0.58/3.1
```

```python
#surface density plots
H, xedges,yedges = histogram2d(J_norm,E, bins=(30,38))
X,Y=meshgrid(yedges[0:-1],xedges[0:-1])
th=4000
print amin(H),amax(H), sum(H[where(H>th)])/sum(H) # this value of 900 selects 67% of the points
levels=array([th])
C=contour(X,Y,H, levels, colors='k', linewidths=2)
```
